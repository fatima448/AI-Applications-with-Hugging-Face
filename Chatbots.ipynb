```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("facebook/blenderbot-400M-distill")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/blenderbot-400M-distill")
```

```python
user_message = "What are the best ways to learn AI?"
inputs = tokenizer(user_message, return_tensors="pt")
output_ids = model.generate(inputs.input_ids, max_length=100)
response = tokenizer.decode(output_ids[0], skip_special_tokens=True)
print(response)
```

```python
def add_message_and_respond(history, new_message):

  history.append({"role":"user", "content":new_message})

  full_conversation = " ".join([message["content"] for message in history])

  inputs = tokenizer(full_conversation, return_tensors="pt")

  output_ids = model.generate(inputs.input_ids, max_length=150, do_sample=True, temperature=0.7 )

  response = tokenizer.decode(output_ids[0], skip_special_tokens=True)

  history.append({"role":"assistant", "content":response})

  return response
```

```python
conversation_history = []

new_message = "What are the best ways to learn AI?"
response = add_message_and_respond(conversation_history, new_message)
print(response)
```
